```python
# pacotes necessários
import imutils
import cv2
from matplotlib import pyplot as plt
```

```python
# carregue a imagem de entrada e mostre suas dimensões, tendo em mente que
# images são representadas como um array NumPy multidimensional com
# shape linhas (altura) x colunas (largura) x canais (profundidade)
# a imagem é carregada nas cores BGR, a linha 6 converte para RGB
image = cv2.imread("decio.jpg")
img = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
(h, w, d) = image.shape
print("width={}, height={}, depth={}".format(w, h, d))
```

```python
# para abrir a imagem em uma janela usar a linha 3 e 4
# na linha 5 e 6 é utilizada a matplotlib para plotar img 
# cv2.imshow("Image", image)
# cv2.waitKey(0)
plt.imshow(img) 
plt.show()
```

```python
# regiões de interesse em uma imgem podem ser selecionadas passando a pos x1,x2 e y1,y2
roi = img[250:390, 250:390]
plt.imshow(roi) 
plt.show()
```

```python
# selecionando apenas 4 pixeis
roi = img[250:252, 220:222]
plt.imshow(roi) 
plt.show()
```

```python
# redimensionamento de imagem
resized = cv2.resize(img, (200, 200))
plt.imshow(resized) 
plt.show()
```

```python
# rotacionando uma imagem utilizando apenas o opencv
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, -45, 1.0)
rotated = cv2.warpAffine(img, M, (w, h))
plt.imshow(rotated) 
plt.show()
```

```python
# rotacionando uma imagem utilizando a imutils
rotated = imutils.rotate(img, -45)
plt.imshow(rotated) 
plt.show()
```

```python
# para rotacionar sem cortar partes da imagem é utilizada a função rotate_bound,
# sua implementação define a angulação contrara a rotate
rotated = imutils.rotate_bound(img, 45)
plt.imshow(rotated) 
plt.show()
```

```python
# aplique um desfoque gaussiano com um kernel 11x11 à imagem para suavizá-la,
# útil ao reduzir o ruído de alta frequência
blurred = cv2.GaussianBlur(resized, (11, 11), 0)
plt.imshow(blurred) 
plt.show()
```

```python
# desenhando retangulo na imagem
output = resized.copy()
imgDraw = cv2.rectangle(output, (60, 75), (110, 140), (0, 255, 0), 2)
plt.imshow(imgDraw) 
plt.show()
```

```python
# ciruclo na iamgem cv2.circle(output, (centro da circunferencia), raio, cor, espessura da linha)
output = resized.copy()
imgDraw = cv2.circle(output, (90, 100), 25,(0, 255, 0), -1)
plt.imshow(imgDraw) 
plt.show()
```

```python
# circulo na imagem
output = resized.copy()
imgDraw = cv2.circle(output, (90, 100), 25,(0, 255, 0), 2)
plt.imshow(imgDraw) 
plt.show()
```

```python
# linha da imagem
output = resized.copy()
imgDraw = cv2.line(output, (50, 50), (150, 150), (0, 255, 0), 2)
plt.imshow(imgDraw) 
plt.show()
```

```python
# escrevendo na imagem
output = resized.copy()
imgDraw = cv2.putText(output, "Decio topzera", (50, 15), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)
plt.imshow(imgDraw) 
plt.show()
```
